```python
data.head()
```

```python
data['Engine HP'].dtype == 'O'
```

```python
data.dtypes
```

```python
for col in data.columns:
    print("dataType:", data[col].dtypes, end=' :')
    if data[col].dtype == 'O':
        print(data[col].nunique())
        print(data[col].unique())
    else:
        print("\n\n")
```

```python
# by analysing the values belongs to object seems to be categorical converting
for col in data.columns:
    if data[col].dtype == 'O':
        data[col] = pd.Categorical(data[col])
```
